```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the first CSV file
data1 = pd.read_csv('file1.csv')

# Read the second CSV file
data2 = pd.read_csv('file2.csv')

# Plot the data from both files
plt.plot(data1['PWM'], data1['RPM'], label='Amsterdam')
plt.plot(data2['PWM'], data2['RPM'], label='Lisbon')

# Add labels and legend
plt.xlabel('RPM')
plt.ylabel('PWM')
plt.title('Motor Output Response Comparison')
plt.legend()

# Show the plot
plt.show()
```
